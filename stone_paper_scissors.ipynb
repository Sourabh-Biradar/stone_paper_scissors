```python
import random
```

```python
def game_begins(player):

    print(f"You : {player}")

    computer = random.choice(['stone','paper','scissors'])
    print(f"Computer : {computer}")
    
    if player == computer:
        print('Draw')
    elif (player == 'stone' and computer == 'paper') or (player == 'paper' and computer == 'scissors') or (player == 'scissors' and computer == 'stone'):
        print('computer wins :(')
    else : 
        print('You win!!!')
```

```python
options = ['stone','paper','scissors']
```

```python
player = input("choose : stone,paper or scissors \n")

if player not in options:
    print('Select vaild option or check spelling')
else :
    game_begins(player)
```
